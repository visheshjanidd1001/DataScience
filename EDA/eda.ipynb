```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')

## on = 'country code' = Merge based on this common column in both DataFrames
## how='left'= Do a left join: keep all rows from df, and match data from df_country where possible
```

```python
final_df.head()
```

```python
## to check datatypes 
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
## pie chart - top 3 country
plt.pie(country_val[:3], labels=country_names[:3],autopct='%1.2f%%')

## country_val[:3]: First 3 values (e.g., top 3 country restaurant counts).
## country_names[:3]: Corresponding country names.
## autopct='%1.1f%%': Shows percentage with 1 decimal point.
```

Observation : Zomato maximum records or transaction from india after that USA and then united kingdoms

```python
final_df.columns
```

```python
ratings = final_df.groupby(["Aggregate rating",'Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating count'})
```

```python
ratings
```

## observation
1. when rating is between 4.5 to 4.9 ---> Excellent
2. when rating are between 4.0 to 4.4 --> very good
3. when ratings are between 3.5 to 3.9 --> good
4. when ratings are between 3.0 to 3.4 --> average 
5. when ratings are between 2.5 to 2.9 ---> average
6. when ratings are between 2.0 to 2.4 ---> poor

```python
ratings.head()
```

## matplotlib.rcParams['figure.figsize'] = (12,6)
matplotlib.rcParams = is a dictionary-like object in Matplotlib that holds all the default configuration settings (also known as "rc settings" or "runtime configuration parameters").

You can view or change plot appearance globally using this object.



```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating count",data = ratings)

```

```python
sns.barplot(x="Aggregate rating",y="Rating count",hue = 'Rating color',data = ratings,palette = ['white','red','orange','yellow','green','green'])
```

## Observation:
1. not rated count is very high
2. maximum number of rating between 2.5 to 3.4

```python
## count plot
sns.countplot(x='Rating color',data = ratings,palette = ['blue','red','orange','yellow','green','green'])
```
