## Black Friday Dataset EDA
Cleaning and preparing dataset for model training

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python

## importing the dataset
df_train = pd.read_csv("train.csv" ,encoding = 'latin-1' )
```

```python
df_train.head()
```

```python
df_test =pd.read_csv("test.csv",encoding= "latin-1")
```

```python
df_test.head()
```

```python
## merge both data 
df = pd.concat([df_train,df_test])
```

```python
df.head()
```

```python
## basic
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
pd.get_dummies(df['Gender'],drop_first=1)
```

```python
## Handling categorical feature GENDER
df['Gender'] = df['Gender'].map({"F": 0,"M": 1})

```

```python
df.head()
```

```python
### Handle categorical feature AGE
df["Age"].unique()
```

```python
df["Age"] = df["Age"].map({'0-17':1, '18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
## fixing 
df_city = pd.get_dummies(df["City_Category"],drop_first=True)
```

```python
df_city.head()
```

```python
df =pd.concat([df,df_city],axis=1)
df.head()
```

```python
## drop city_category
df.drop("City_Category",axis=1,inplace=True)
```

```python
df.head()
```

```python
## Missing Values
df.isnull().sum()
```

```python
df['Product_Category_1'].unique()
```

```python
## Focus on replacing missing values 
df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].value_counts()
```

```python
df['Product_Category_2'].mode()[0]
```

```python
##Replace the missing value with mode 
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'])
```

```python
df['Product_Category_2'].isnull().sum()
```
